```python
import numpy as np
from scipy.optimize import linprog
```

Цільова функція (дохід у тис. $):
maximize Z = 2*x1 + 9*x2 + 6*x3

```python
c = [-2, -9, -6]
```

```python
A = [
    [12,  6,  2],
    [12, 24, 18],
    [12, 18, 12]
]
b = [320, 192, 180]
```

Обмеження:
12*x1 + 6*x2 + 2*x3  <= 320   (дизайнери)
12*x1 +24*x2 +18*x3 <= 192   (розробники)
12*x1 +18*x2 +12*x3 <= 180   (тестувальник)

```python
bounds = [(0, None), (0, None), (0, None)]
```

```python
res = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')
```

```python
res.message
```

```python
x1, x2, x3 = res.x
print(f" Оптимальна кількість:")
print(f"  Сайти (x1):               {x1:.4f}")
print(f"  Інтернет-магазини (x2):   {x2:.4f}")
print(f"  Інтеграції ERP (x3):      {x3:.4f}")
```

```python
income = 2*x1 + 9*x2 + 6*x3
print(f"\n Максимальний дохід: {income:.2f} тис. $")
```
